# Comparison of classic projections

```python
# See http://michel.thoby.free.fr/Fisheye_history_short/Projections/Models_of_classical_projections.html

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tck

theta = np.linspace(0,np.pi)

fig, ax = plt.subplots(figsize=(5,5))

ax.plot((theta/np.pi)[theta<np.pi/2], np.tan(theta)[theta<np.pi/2], c="r", label="Perspective")
ax.plot((theta/np.pi), 2.*np.tan(theta/2.), c="c", label="Stereographic")
ax.plot((theta/np.pi), theta, c="g", label="Equidistant")
ax.plot((theta/np.pi), 2*np.sin(theta/2.), c="orange", label="Equisolid")
ax.plot((theta/np.pi)[theta<np.pi/2], np.sin(theta)[theta<np.pi/2], c="b", label="Ortographic")

ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
ax.xaxis.set_major_locator(tck.MultipleLocator(base=0.25))

plt.legend(loc="upper right")
plt.grid()
plt.xlim([0,1])
plt.ylim([0,5])
plt.title("Radius-angle relation")
plt.xlabel(r"$\theta$",fontsize=14)
plt.ylabel(r"$r/f$",fontsize=14)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot((theta/np.pi)[theta<np.pi/2], 1./(np.cos(theta)**3.)[theta<np.pi/2], c="r", label="Perspective")
ax.plot((theta/np.pi)[theta<np.pi], 1./(np.cos(theta/2.)**4.)[theta<np.pi], c="c", label="Stereographic")
ax.plot((theta/np.pi)[theta<np.pi], (theta/np.sin(theta))[theta<np.pi], c="g", label="Equidistant")
ax.plot((theta/np.pi)[theta<np.pi], np.ones((theta/np.pi)[theta<np.pi].shape), c="orange", label="Equisolid")
ax.plot((theta/np.pi)[theta<np.pi/2], np.cos(theta)[theta<np.pi/2], c="b", label="Ortographic")

ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
ax.xaxis.set_major_locator(tck.MultipleLocator(base=0.5))

plt.legend(loc="upper right")
plt.grid()
plt.xlim([0,1])
plt.ylim([0,10])
plt.title("Area distortion")
plt.xlabel(r"$\theta$",fontsize=14)
plt.ylabel(r"$dA_{image}/dA_{sphere}$",fontsize=14)
plt.show()
```
